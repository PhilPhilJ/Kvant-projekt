```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sqrt, pi, sin, cos
import math
from scipy import special
%matplotlib widget
```

### Making a plot for the harmonic oscillator.

The harmonic oscillator has wave functions:

$\psi_n(x) = \frac{1}{\sqrt{n}} (\hat{a}_+)^n \psi_0$

From the potential:

$V(x) = \frac{1}{2} m \omega^2 x^2 $


with energies:

$E_n=(n + \frac{1}{2})\hbar \omega$

keeping it simple, let $\omega = 1$

```python
hbar = 1
m = 1
omega = 1
sf = 10000
x = np.linspace(-10,10,sf)
zeta = sqrt(omega)*x
H = [1, 2*zeta, 4*zeta**2-2, 8*zeta**3-12*zeta, 16*zeta**4 - 48*zeta**2 + 12, 32*zeta**5 - 160*zeta**3 + 120*zeta]
```

```python
def psiS_n(n_2,x):
    return (1/pi)**(1/4) * 1/(sqrt(2**(n_2))*math.factorial(n_2))*H[n_2]*np.exp((-x**2)/2)
```

```python
def EH_n(n_2):
    return (n_2 + 1/2)*hbar*omega
```

```python
plt.close()
fig, ax = plt.subplots(1, 3, sharex=False, sharey=True)

plot_width = 5

omegas = np.array([1,1.2,1.4])

for j in range(len(omegas)):
    for i in range(5): 
        
        omega = omegas[j]
        x = np.linspace(-plot_width, plot_width, sf)

        ax[j].hlines(y=EH_n(i), xmin=-plot_width, xmax=plot_width, linewidth=2, color='black',linestyle='solid', alpha=0.2)
        ax[j].plot(x, psiS_n(i,x) + EH_n(i))
        ax[j].title.set_text(r'$\omega$' + "= " + str(omega))

    #creating the potential manually
    
    

    ax[j].plot(x,1/2*m*omega**2*x**2, color ="black", linewidth=3)
    ax[j].set_ylim([-0.5, 10])

fig.text(0.5, 0.04, 'position (x)', ha='center')
fig.text(0.04, 0.5, 'Energier', va='center', rotation='vertical')
```
