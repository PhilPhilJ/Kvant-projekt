## Making plots for the harmonic oscillator and infinite square well

### Making a plot for the infinete squre well.

The infinite square well has wave functions:

$\psi_n(x) = \sqrt{\frac{2}{a}}sin(\frac{n\pi}{a}x)$ 

From the potential:

\
 $V(x) =
  \begin{cases} 
      \hfill 0,    \hfill & 0 \le x \le a \\
      \hfill \infty, \hfill & otherwise \\
  \end{cases}$


with energies:

$E_n=\frac{n^2\pi^2\hbar^2}{2ma^2}$

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sqrt, pi, sin, cos
%matplotlib widget
```

```python
hbar = 1
m = 1
start_n = 1
max_n = 5
ex_amp = 10
```

```python
def psiS_n(n,x):
    return ex_amp*sqrt(2/a)*sin(n*pi*x/a)
```

```python
def ES_n(n):
    return (n*pi*hbar)**2/(2*m*a**2)
```

```python
sf = 10000
n = np.arange(start_n, max_n+1, 1, dtype=int)
```

```python
plt.close()

aes = np.array([1,1.1,1.2])

fig, ax = plt.subplots(1, 3, sharex=False, sharey=True)

for j in range(len(aes)):
    for i in range(len(n)): 
        
        a = aes[j]
        x = np.linspace(0,a,sf)

        ax[j].hlines(y=ES_n(n[i]), xmin=0, xmax=a, linewidth=2, color='black',linestyle='solid', alpha=0.2)
        ax[j].plot(x,psiS_n(x,n[i]) + ES_n(n[i]))
        #ax[j].set_ylabel("Brndlngde")
        ax[j].title.set_text('a = ' + str(aes[j]))

        #creating the potential manually
        ax[j].vlines(x=0, ymin=0, ymax=140, color ="black", linewidth=4)
        ax[j].vlines(x=a, ymin=0, ymax=140, color ="black", linewidth=4)
        ax[j].hlines(y=0, xmin=0, xmax=a, color ="black", linewidth=4)

fig.text(0.5, 0.04, 'Brnlngde', ha='center')
fig.text(0.04, 0.5, 'Energier', va='center', rotation='vertical')
```

```python
plt.close()
```
